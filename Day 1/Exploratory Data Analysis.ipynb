```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df
```

```python
final_df.shape
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_name = final_df.Country.value_counts().index
country_name
```

```python
country_value = final_df.Country.value_counts().values
country_value
```

```python
# Pie Chart- top 5 country
plt.pie(country_value[:3], labels=country_name[:3], autopct="'%1.2f%%'")
```

Zomato maximum transections are from India.

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()\
.rename(columns={0:"Rating Count"})
ratings
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x = "Aggregate rating", y = "Rating Count", data = ratings)
```
